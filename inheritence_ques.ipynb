```python
# #8 Multiple Methods

# Parent class Bank with deposit(). Child class Account with withdraw().

class Bank:
    def info(self,name,age,balance):
        self.name=name
        self.age=age
        self.balance=balance
    def deposit(self,dp):
        self.dp=dp
        self.balance+=self.dp
        print(f"Deposit of {self.dp} Success. Updated balance: {self.balance}")
        
class Account(Bank):
    def withdraw(self,wd):
        self.wd=wd
        self.balance-=self.wd
        print(f"Withdrawl of {self.wd} Success. Updated balance: {self.balance}")

obj=Account()
obj.info("XYZ",12,1000)
obj.deposit(1000)
obj.withdraw(2000)
```

```python
# #11 Basic Multiple Inheritance

# Create classes Father and Mother. Child class Child inherits both.

class Father:
    def ft(self,fpocketmoney):
        self.fpocketmoney=fpocketmoney
class Mother:
    def mt(self,mpocketmoney):
        self.mpocketmoney=mpocketmoney
class Child(Father,Mother):
    def totalpm(self):
        print(f"Total pocket money child have: {(self.fpocketmoney+self.mpocketmoney)}")
c = Child()
c.ft(1000)
c.mt(1000)
c.totalpm()       
```

```python
# #27 Method Resolution Order (MRO) -> Searching order btata hai

# Create multiple inheritance and print MRO.

class A:
    def show(self):
        print("A")
class B:
    def show(self):
        print("B")
class C(A,B): #Searching order C->A->B
    pass        
class D(B,A): #Searching order C->B->A
    pass        
obj = C()
obj.show()
obj1 = D()
obj1.show()
```

```python
# #28 super() in Multiple Inheritance

# Use super() correctly in multi-level classes.

class A:
    def __init__(self):
        print("Hellooo from A")
class B(A):
    def __init__(self):
        super().__init__()
        print("Hellooo from B")
class C(B):
    def __init__(self):
        super().__init__()
        print("Hellooo from C")
obj = C()
```

```python
# #29 Dynamic Method Call

# Call parent method using object of child dynamically.

class A:
    def call(self,msg):
        self.msg=msg
        print(self.msg)
class B(A):
    def __init__(self):
        print("Default msg")
obj = B()
obj.call("Parent method called using object of child dynamically")        
```

```python
# #32 Library System

# Class Book, inherited by IssuedBook with extra features.
class Book:
    def book(self,bname,daycharge):
        self.bname=bname
        self.daycharge=daycharge
class IssuedBook(Book):
    def issue(self,name,days):
        self.name=name
        self.days=days
    def show(self):
        if self.days > 7:
            print(f"Book name: {self.bname} issued to {self.name} for {self.days} days.")
            print(f"Penalty fine for not returning book under 7 days: Rs{(self.days-7)*10}(extra).\nFinal Charge: RS{(self.daycharge*self.days)+(self.days-7)*10}\n")  
        else:
         print(f"Book name: {self.bname} issued to {self.name} for {self.days} days.\nFinal Charge: RS{self.daycharge*self.days}\n")    
b1 = IssuedBook()
b1.book("any book",20)
b1.issue("Ayush",8)  
b1.show()
b1 = IssuedBook()
b1.book("any book",20)
b1.issue("Ayush",5)  
b1.show()           
```
