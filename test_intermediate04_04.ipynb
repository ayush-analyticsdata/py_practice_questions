```python
# 9. Print pattern:


#    1
#    12
#    123
#    1234


#    Use nested for loops.

for i in range(1,5):
  for j in range(1,i+1):
    print(j,end="")
  print()
```

```python
# 10. Find the factorial of a number using while loop.
num=int(input("Enter a number to find factorial: "))
factorial=1
i = 1
while i<=num:
  factorial=factorial*i
  i+=1
print(f"{factorial} is a factorial of {num}")
```

```python
# 11. Count frequency of vowels and consonants in a string using loop and condition.

sen = input("Enter a sentence: ")
vowel = 0
consonants=0

for i in sen:
  if i == "a" or i == "A":
    vowel+=1
  elif i == "e" or i== "E":
    vowel+=1
  elif i == "i" or i== "I":
    vowel+=1
  elif i == "o" or i== "O":
    vowel+=1
  elif i == "u" or i== "U":
    vowel+=1
  elif i == " ":
    continue
  else:
    consonants=consonants+1

print(f"Vowels count: {vowel} and Consonants count: {consonants}")

```

```python
# 12. Take a list of numbers:

# Print only positive even numbers
# Skip negative numbers using continue

li = [1,2,3,4,5,-1,-2,-4,-2,10,12,68,96,31,33,35]

for i in li:
  if i<0:
    continue
  if i%2==0:
    print(i)
```

```python
# 13. Find the largest and smallest number in a list using loop.

li = [4,5,3,8,9,11,2,1,7]
largest = li[1]
smallest = li[1]

for i in li:
  if i>largest:
    largest = i
  elif i<smallest:
    smallest = i
print("Largest element is:",largest)
print("Smallest elememeent is:",smallest)
```

```python
# 14. Print all Armstrong numbers between 1 to 1000.

for num in range(1,1001):

  orignal = num
  power = len(str(num))
  sum = 0

  while orignal > 0:
    digit = orignal%10
    sum += digit ** power
    orignal = orignal//10

  if num == sum:
    print(num)
```

```python
# 15. Take a number and check if it is perfect number.

# examples of perfect number = 6,28,496 or etc...

num = int(input("Enter a number: "))
divisor = 0
for i in range(1,num):
  if num%i==0:
    divisor=divisor+i
if divisor == num:
  print(f"{num} is a perfect number")
else:
  print(f"{num} is not a perfect number")


```

```python
# 16. Create a menu-driven program using while loop:

# 1 → Add two numbers
# 2 → Subtract
# 3 → Exit

menu = int(input("1.Addition of two numbers. \n2.Substraction of two numbers. \n3.Exit.\n"))
i = 1

while menu == 1:
  num1=int(input('Enter 1st number: '))
  num2=int(input('Enter 2nd number: '))
  add = num1+num2
  print(f"Addition of {num1} and {num2} is {add}.")
  i+=1
  break

while menu == 2:
  num1=int(input('Enter 1st number: '))
  num2=int(input('Enter 2nd number: '))
  subs = num1-num2
  print(f"Substraction of {num1} and {num2} is {subs}.")
  i+=1
  break

while menu == 3:
  print("Successfully Exited")
  i+=1
  break
```

```python
# 17. Print numbers from 1 to 100:

# Skip multiples of 3
# Stop loop if number > 80

for i in range(1,101):
  if i%3==0:
    continue
  elif i>80:
    break
  else:
    print(i)
```
