```python
# 16. Flatten a 2D list
#     Input: [[1, 2], [3, 4], [5]]
#     Output: [1, 2, 3, 4, 5]

a =[[1,2],[3,4],[5]]
flatten = list()

for i in a:
    if type(i)==list:
        for j in i:
            flatten.append(j)
    else:
        flatten.append(i)
print(flatten)                
```

```python
# 17. Create a multiplication table (1 to 3)
#     Output: [1, 2, 3, 2, 4, 6, 3, 6, 9]

a = [1,2,3]
b=list()
for i in a:
    for j in range(1,4):
     b.append(i*j)
print(b)
```

```python
# 18. Create pairs from two lists
#     Input: [1, 2], ['a', 'b']
#     Output: [(1, 'a'), (1, 'b'), (2, 'a'), (2, 'b')]

a = [1,2]
b = ["a","b"]
c = list()
for i in range(len(a)):
    for j in range(len(b)):
        c.append((a[i],b[j]))
print(c)
           
```

```python
# 19. Flatten list but keep only even numbers
#     Input: [[1, 2], [3, 4], [5, 6]]
#     Output: [2, 4, 6]

a = [[1, 2], [3, 4], [5, 6]]
even = list()

for i in a:
    if type(i)==list:
        for j in i:
            if j%2==0:
                even.append(j)
print(even)                
```

```python
# 21. Find common elements between two lists
#     Input: [1, 2, 3], [2, 3, 4]
#     Output: [2, 3]

a = [1,2,3]
b = [2,3,4]

common = [i for i in a for j in b if i==j]
print(common)
```

```python
# 25. Create list of prime numbers from 1 to 20
#     Output: [2, 3, 5, 7, 11, 13, 17, 19]
               
a = list()

for num in range(1,21):
    for i in range(2,num):
     if num%i == 0:
        break
    else:
        a.append(num)
print(a)       
```

```python
# 26. Matrix transpose using list comprehension
#     Input: [[1, 2, 3], [4, 5, 6]]
#     Output: [[1, 4], [2, 5], [3, 6]]

a= [[1, 2, 3], [4, 5, 6]]
b=list()
c=list()

for i in a:
    if type(i)==list:
        for j in i:
            b.append(j)
            
for k in range(0,3):
    c.append(([b[k],b[k+3]]))
    
print(c)            

```

```python
# 27. Flatten 3D list
#     Input: [[[1, 2], [3]], [[4, 5]]]
#     Output: [1, 2, 3, 4, 5]

a=[[[1, 2], [3]], [[4, 5]]]
b=list()

for i in a:
        for j in i:
                for k in j:
                    b.append(k)
print(b)                    

```

```python
# 28. Create list of factorial values from 1 to 5
#     Output: [1, 2, 6, 24, 120]

a = []

for num in range(1,6):
    fact = 1
    for i in range(1,num+1):
        fact*=i
    a.append(fact)
    
print(a)            




```

```python
# 30. Generate list of all substrings of a string
#     Input: "abc"
#     Output: ["a", "ab", "abc", "b", "bc", "c"]
a = "abc"
b = list()

for i in range(len(a)):
    b.append(a[i])
    for j in range(0,len(a)-1):
        if a[j]+a[j+1] not in b :
         b.append(a[j]+a[j+1])
b.append(a)         
print(b)
```
