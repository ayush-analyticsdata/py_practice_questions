```python
# 18. Take a list and:

# Count even, odd, positive, negative numbers using one loop

a = [1,2,3,-4,5,6,-7,8,9]
even = 0
odd = 0
positive = 0
negative = 0
for i in a:
  if i%2==0 and i > 0:
    positive+=1
    even+=1
  elif i%2==0 and i < 0:
    negative+=1
    even+=1
  elif i%2!=0 and i > 0:
    positive+=1
    odd+=1
  elif i%2!=0 and i < 0:
    negative+=1
    odd+=1
  else:
    negative+=1
print(f"Even number count: {even}")
print(f"Odd number count: {odd}")
print(f"Positive number count: {positive}")
print(f"Negative number count: {negative}")


```

```python
# 19. Print a pyramid pattern:

#    *
#   ***
#  *****
n=3
for i in range(1,n+1):
    for j in range(n-i):
       print(" ",end="")
    for k in range(2*i-1):
       print("*",end="")
    print()
```

```python
# 21. Find second largest number in a list using loop (without sorting).
a=[1,2,3,4,5]
max = a[0]

for i in a:
    if i>max:
        max = i
a.remove(max)
s_max = a[0]
for j in a:
    if j>s_max:
        s_max = j
print(f"Second max: {s_max}")
```

```python
# 22. Simulate ATM system:

# Balance check
# Deposit
# Withdraw (check insufficient balance using if-else)

balance = 5000
o= int(input("Enter your preffered option:\n 1.Balance Check \n 2.Deposit \n 3.Withdraw \n 4.Exit"))
if o==1:
    print(f"Balance: {balance}")
elif o==2:
    deposit = int(input("Enter amount to deposit:"))
    balance+=deposit
    print(f"New balance: {balance}")
elif o==3:
    withdraw = int(input("Enter amount to withdraw:"))
    if withdraw<=balance:
     balance-=withdraw
     print(f"Withdrawl successful of Rs{withdraw}! New balance: {balance}")
    else:
        print("Insufficient Balance") 
elif o==4:
    print("Menu exit successfully")

else:
    
    print("Invalid choice")    
```

```python
# 23. Check if a string is palindrome ignoring spaces and case.
import string
a = "N a M a n"
new=""
rev = ""
for i in a:
    if i.isalnum():
       new+=i.lower()
    else:
        continue
for j in new:
    rev = j+rev
if rev==new:
    print("Palindrome")
else:
    print("Not Palindrome")        
```

```python
# 25. Process a list of student marks:

# Print grade:

#   * ≥90 → A
#   * ≥75 → B
#   * ≥50 → C
#   * else → Fail
# Also count how many students passed/failed

a = [95,49,64,40,82,75]
passed=0
failed=0

for i in a:
    if i >=90:
        print(f"Marks:{i} Grade A")
        passed+=1
    elif i >=75:
        print(f"Marks:{i} Grade B")
        passed+=1
    elif i >=50:
        print(f"Marks:{i} Grade C")
        passed+=1
    else:
        print(f"Marks:{i} Failed")
        failed+=1
print("Passed:",passed)
print("Failed:",failed)

```
