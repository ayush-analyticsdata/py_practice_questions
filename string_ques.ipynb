```python
# 1. Print each character of a string | Input: "hello" | Output: h e l l o

a="hello"
for i in a:
    print(i)
```

```python
# 2. Count total characters (without using len) | Input: "python" | Output: 6
a = "python"
count = 0
for i in a:
    count+=1
print(count)    
```

```python
# 3. Count occurrence of a character | Input: "banana", 'a' | Output: 3

a = "banana"
count = 0
for i in a:
    if i=="a":
        count+=1
print(count)        
```

```python
# 4. Print characters at even index positions | Input: "abcdef" | Output: a c e

a="abcdef"

for i in range(len(a)):
    if i%2==0:
        print(a[i],end=' ')
    
```

```python
# 5. Print characters at odd index positions | Input: "abcdef" | Output: b d f

a="abcdef"

for i in range(len(a)):
    if i%2==0:
        continue
    else:
        print(a[i],end=' ')
```

```python
# 6. Reverse a string (without slicing) | Input: "hello" | Output: olleh

a="hello"
rev = ""

for i in a:
    rev = i+rev
print(rev)    
```

```python
# 7. Count vowels | Input: "education" | Output: 5

a="education"
vowel=0
for i in a:
    if i =="a" or i =="e" or i =="i" or i =="o" or i =="u":
        vowel+=1
print(vowel)        
```

```python
# 8. Count consonants | Input: "hello" | Output: 3

a="hello"
con = 0
 
for i in a: 
 if i != "a" and i != "e"and i != "i" and i != "o" and i != " u":
   con+=1
print(con)  
```

```python
# 9. Count digits in a string | Input: "a1b2c3" | Output: 3

a="a1b2b3"
digit=0

for i in a:
    if i.isnumeric():
        digit+=1
print(digit)        
```

```python
# 10. Count spaces in a string | Input: "hi hello how" | Output: 2

a="hi hello how"
count=0

for i in a:
    if i == " ":
        count+=1
print(count)        
```

```python
# 11. Replace spaces with '-' | Input: "hi hello" | Output: hi-hello

a = "hi hello"
b = a.replace(" ","-")

print(b)
```

```python
# 12. Find first occurrence index of a character | Input: "banana", 'a' | Output: 1

s ="banana"

for i in range(len(s)):
    if s[i] == "a" :
        print("First occurence Index of a:",i)
        break

```

```python
# 13. Find all indexes of a character | Input: "banana", 'a' | Output: 1 3 5

z="banana"

for i in range(len(z)):
    if z[i] == "a":
        print(i,end=" " )

```

```python
# 14. Check if string is palindrome | Input: "madam" | Output: Palindrome

og = "madam"
rev = ""

for i in og:
    rev = i + rev
if og == rev:
    print("Yes it is a palindrome")
else:
    print("Not a palindrome")            
```

```python
# 15. Find length of string manually | Input: "data" | Output: 4

a="data"
count=0
for i in a:
    count+=1
print("Length:",count)    
```

```python
# 16. Count uppercase letters | Input: "HeLLo" | Output: 3
import string 

a="HeLLo"
b = string.ascii_uppercase
count=0
for i in a:
    for j in b:
        if i==j:
            count+=1
print(f"Upper case are: {count}")            
```

```python
# 17. Count lowercase letters | Input: "HeLLo" | Output: 2
import string 

a="HeLLo"
b = string.ascii_lowercase
count=0
for i in a:
    for j in b:
        if i==j:
            count+=1
print(f"Lower case are: {count}") 
```

```python
# 19. Remove all digits | Input: "a1b2c3" | Output: abc

a="a1b2c3"
new=""

for i in a:
    if i.isnumeric():
        continue
    else:
        new=new+i
print(new)        
```

```python
# 20. Print only unique characters | Input: "programming" | Output: progamin

a="programming"
unique=""

for i in a:
    if i not in unique:
        unique+=i
print(unique)
```

```python
# 24. Check if string contains only alphabets | Input: "hello" | Output: Valid

a="hello"
valid = True
for i in a:
    if i.isdigit():
        valid = False
        break
if valid==True:
    print("Valid") 
else:
    print("Not Valid")       
```

```python
# 25. Find longest repeating character sequence | Input: "aaabbccccd" | Output: cccc

a="aaabbccccd"
rep_a = 0
rep_b = 0
rep_c = 0
rep_d = 0

for i in a:
    if i =="a":
        rep_a+=1
    elif i =="b":
        rep_b+=1
    elif i =="c":
        rep_c+=1
    elif i =="d":
        rep_d+=1
if rep_a>rep_b and rep_a>rep_c and rep_a>rep_d:
    print("a is longest repeating")
elif rep_b>rep_a and rep_b>rep_c and rep_b>rep_d:
    print("b is longest repeating")
elif rep_c>rep_b and rep_c>rep_a and rep_c>rep_d:
    print("c is longest repeating")
elif rep_d>rep_b and rep_d>rep_c and rep_d>rep_a:
    print("d is longest repeating")
    
```

```python
# 26. Shift each character by 1 | Input: "abc" | Output: bcd

import string
a="bcd"
b = string.ascii_lowercase
c=""

for i in range(len(a)):
    for j in range(len(b)):
        if a[i]==b[j]:
            c=c+b[j+1]
print(c)            
```

```python
# 27. Convert lowercase to uppercase manually | Input: "abc" | Output: ABC

z="abc"
b=""
for i in z:
    b = b+chr(ord(i)-32)
print(b)

# for uppercase to lowercase

y="ABC"
c=""
for i in y:
    c = c+chr(ord(i)+32)
print(c)
```
