```python
# Check whether input number is even or odd. Handle invalid input using exception handling.

try:
    x=eval(input("Enter any number: "))
    if x%2==0:
     print("Number is even")
    else:
     print("Number is odd")
except:
    print("Invalid input. Enter integers only")    
else:
    print("Successfully Evaluated")
finally:
    print("Exit")

   
```

```python
# Take a number from user. If input is valid: (using math build-in function)
# a.Print ceil if number > 0
# b.Print floor if number < 0
# c.Handle invalid input

import math
try:
    num=eval(input("Enter any number"))
    if type(num) is str:
        raise TypeError("Invalid input. Please enter integer only")
    if num>0:
        a=math.ceil(num)
        print("Ceil value is:",a)
    elif num<0:
        b=math.floor(num)
        print("Floor value is:",b)

except TypeError as t:
    print("Exception Handled:",t)
except Exception as e:
    print(e)
else:
    print("Successfully Evaluated")
finally:
    print("Exit")

```

```python
# Write a program to divide two numbers entered by user. Handle the case where the denominator is zero.
# Print the result if no error is found. using try-except-else-finally

try:
    a=eval(input("Enter 1st number:"))
    b=eval(input("Enter 2nd number:"))
    
    c=a/b
    print('Result:',c)
except ZeroDivisionError:
    print("Denominator cannot be zero")
except Exception as e:
    print("Error: ",e)
else:
    print("Operation Done Successfully")
finally:
    print("Exit time")
    
```

```python
# Write a program to convert a string input into an integer. Handle invalid input.

try:
    x=input("Enter a number")
    x=int(x)
    print("Done!",type(x))
except:
    print("Please enter interger input only")
else:
    print("Successfully converted string into integer")
finally:
    print("Byy")
```
