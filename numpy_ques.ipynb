# 🔹 Logical / Matrix Questions


```python
#28

# Create a 3x3 matrix using NumPy.

import numpy as np

arr3x3 = np.array([[1,2,3],[1,2,3],[1,2,3]])
print(arr3x3)
```

```python
#29

# Transpose a matrix.

# Input:

# [[1,2],
#  [3,4]]

# Output:

# [[1 3]
#  [2 4]]

import numpy as np

a = np.array([[1,2],
             [3,4]])
b = np.transpose(a)
print(b)
```

```python
#30

# Find determinant of matrix.

# Input:

# [[1,2],
#  [3,4]]

# Output:
# -2
import numpy as np
a = np.array([[1,2],
              [3,4]])

b = np.linalg.det(a)
print(b)
```

```python
#31
#  perform matrix multiplication.

# Input:
# Matrix A = [[1,2]]
# Matrix B = [[3],[4]]

# Output:
# [[11]]

import numpy as np
arr1 = np.array([[1,2]])
arr2 = np.array([[3],[4]])

arr3 = np.matmul(arr1,arr2)
print(arr3)
```

```python
#32

# Find eigenvalues of matrix.

# Input:

# [[2,0],
#  [0,3]]

# Output:
# [2,3]
import numpy as np

arr = np.array([[2,0],[0,3]])
eg = np.linalg.eigvals(arr)
print(eg)

```

```python
#33
# Reshape array into 2x2 matrix.

# Input: [1,2,3,4]
# Output:

# [[1 2]
#  [3 4]]

import numpy as np
a = np.array([1,2,3,4])
c = a.reshape(2,2)
print(c)
```

```python
#34
# Flatten a multidimensional array.

# Input:

# [[1,2],
#  [3,4]]

# Output:
# [1 2 3 4]

import numpy as np
a = np.array( [[1,2],
                [3,4]])
b = a.reshape(-1)
print(b)
```

```python
# #35

# Concatenate two NumPy arrays.

# Input: [1,2], [3,4]
# Output: [1 2 3 4]

import numpy as np

arr1 = np.array([1, 2])

arr2 = np.array([3, 4])

arr = np.concatenate((arr1, arr2))

print(arr)
```
