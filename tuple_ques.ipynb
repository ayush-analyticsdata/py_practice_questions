```python
# 1. Print all elements of a tuple using a for loop
#    Input: (1, 2, 3, 4)
#    Output: 1 2 3 4

a = (1,2,3,4)

for i in a:
    print(i,end=" ")
```

```python
# 2. Print all elements using while loop
#    Input: (5, 6, 7)
#    Output: 5 6 7


a=(5,6,7)
i=1
while i:
   print(a)
   i+=1
   break
```

```python
# 3. Find length of tuple (without using len)
#    Input: (10, 20, 30)
#    Output: 3

a = (10,20,30)
count = 0

for i in a:
    count+=1

print(count)    
```

```python
# 4. Find sum of all elements
#    Input: (1, 2, 3, 4)
#    Output: 10


a=(1,2,3,4)
sum=0

for i in a:
    sum+=i
print(sum)    
```

```python
# 5. Find maximum element
#    Input: (3, 7, 2, 9, 5)
#    Output: 9

a = (3,7,2,9,5)
max = a[0]

for i in a:
    if i>max:
        max = i
print(max)        
```

```python
# 6. Find minimum element
#    Input: (3, 7, 2, 9, 5)
#    Output: 2
a = (3,7,2,9,5)
min = a[0]

for i in a:
    if i<min:
        min = i
print(min)
```

```python
# 7. Count occurrence of a number
#    Input: Tuple = (1, 2, 2, 3, 2), Number = 2
#    Output: 3

a = (1,2,2,3,2)
num = eval(input("Enter num to check occurence"))
count = 0

for i in a:
    if i == num:
        count+=1
print(f"Occurence of num {num} is {count}")        

```

```python
# 8. Create a new tuple by adding an element
#    Input: (1, 2, 3), Add = 4
#    Output: (1, 2, 3, 4)

a=(1,2,3)
b = list(a)
b.append(4)
b=tuple(b)
print(b)  
```

```python
# 9. Insert an element at specific index
#    Input: Tuple = (1, 2, 4), Element = 3, Index = 2
#    Output: (1, 2, 3, 4)

a = (1,2,4)
b = list(a)
b.insert(2,3)
b=tuple(b)
print(b)
```

```python
# 15. Remove duplicates from tuple
#     Input: (1, 2, 2, 3, 1)
#     Output: (1, 2, 3)

a = (1,2,2,3,1)
b = list()

for i in a:
    if i not in b:
        b.append(i)
b = tuple(b)
print(b)        
```

```python
# 16. Sort tuple in ascending order
#     Input: (5, 2, 8, 1)
#     Output: (1, 2, 5, 8)

input = (5,2,8,1)
b = list(input)
b.sort()
b = tuple(b)
print(b)    
        
```

```python
# 17. Sort tuple in descending order
#     Input: (5, 2, 8, 1)
#     Output: (8, 5, 2, 1)
input = (5,2,8,1)
b = list(input)
b.sort(reverse=True)
b = tuple(b)
print(b) 
```

```python
# 18. Find second largest element
#     Input: (10, 20, 5, 30)
#     Output: 20

a = (10,20,5,30)
b = list(a)
b.sort()
print("Second largest:",b[-2])
```

```python
# 19. Check if element exists
#     Input: Tuple = (1, 2, 3), Element = 2
#     Output: Found

a = (1,2,3)
b = list(a)
for i in b:
 if 2 in b:
    print("Found")
    break
 else:
    print("Not found") 
    break 
 
```

```python
# 20. Count even numbers
#     Input: (1, 2, 3, 4, 5, 6)
#     Output: 3

a = (1,2,3,4,5,6)
b = list(a)
even=0
for i in b:
    if i%2==0:
        even+=1
print("Even number count: ",even)       
```

```python
# 21. Count odd numbers
#     Input: (1, 2, 3, 4, 5)
    # Output: 3


a = (1,2,3,4,5,)
b = list(a)
odd=0
for i in b:
    if i%2!=0:
        odd+=1
print("Odd number count: ",odd)
```

```python
# 22. Create new tuple with squares of elements
#     Input: (1, 2, 3)
#     Output: (1, 4, 9)

a = (1,2,3)
a = list(a)
b = [i*i for i in a]
b=tuple(b)
print(b)

```

```python
# 23. Remove all negative numbers
#     Input: (1, -2, 3, -4, 5)
#     Output: (1, 3, 5)

a = (1,-2,3,-4,5)
a = list(a)

b = [i for i in a if i>0]
b = tuple(b)
print(b)
```

```python
# 24. Find index of an element (without using index())
#     Input: Tuple = (10, 20, 30), Element = 20
#     Output: 1

a = (10,20,30)
b = list(a)
for i in range(len(b)):
    if b[i]==20:
        print("Index of 20 is",i)
    else:
        continue    
```

```python
# 25. Rotate tuple right by 1 position
#     Input: (1, 2, 3, 4)
#     Output: (4, 1, 2, 3)

a = (1,2,3,4)
a = list(a)
b = list()

for i in range(len(a)-1,-1,-1):
    b.append(a[i])
    break
for j in range(0,len(a)-1):
    b.append(a[j])
print(b)

   
```

```python
# 26. Find common elements between two tuples
#     Input: (1, 2, 3), (2, 3, 4)
#     Output: (2, 3)

a =(1,2,3)
b = (2,3,4)
c = list()
a = list(a)
b=list(b)

for i in a:
    for j in b:
        if i == j:
            c.append(i)
c = tuple(c)            
print(c)            
```

```python
# 27. Flatten a nested tuple (1 level)
#     Input: (1, (2, 3), 4)
#     Output: (1, 2, 3, 4)

a=(1,(2,3),4)
a=list(a)
b=list()

for i in a:
    if type(i)==tuple:
        for j in i:
            b.append(j)
    else:
        b.append(i)   
b=tuple(b)
print(b)             
```

```python
# 28. Remove element at specific index
#     Input: Tuple = (1, 2, 3, 4), Index = 2
#     Output: (1, 2, 4)

a =(1,2,3,4)
a=list(a)

b=[i for i in a if i!=a[2]]
b=tuple(b)
print(b)
```

```python
# 29. Count frequency of each element
#     Input: (1, 2, 2, 3, 1)
#     Output: {1:2, 2:2, 3:1}

a = (1,2,2,3,1)
a = list(a)
c1=0
c2=0
c3=0
for i in a:
    if i==1:
        c1+=1
    elif i==2:
        c2+=1
    elif i==3:
        c3+=1
output = {1:c1,2:c2,3:c3}
print(output)                       
```

```python
# 30. Check if tuple is palindrome
#     Input: (1, 2, 3, 2, 1)
#     Output: Palindrome

a = (1,2,3,2,1)
a = list(a)
b = list()
for i in range(len(a)-1,-1,-1):
    b.append(a[i])

if a==b:
 print("Palindrome")
else:
 print("Not a palindrome")    
```
