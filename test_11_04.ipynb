```python
# #1 Reverse a String
# Write a program to reverse a given string without using built-in reverse functions.
# Input: "hello"
# Output: "olleh"

a = "hello"
rev = ""

for i in a:
    rev = i+rev
print(rev)    
```

```python
# #2 Check Palindrome
# Write a program to check whether a given string is a palindrome.
# Input: "madam"
# Output: True

a = "madam"
rev=""
for i in a:
    rev=i+rev
if rev==a:
    print("palindrome")
else:
    print("not a palindrome")        
```

```python
# #3 Count Vowels
# Write a program to count the number of vowels in a string.
# Input: "programming"
# Output: 3

a="programming"
vowel = 0

for i in a:
    if i =="a" or i =="e" or i =="i" or i =="o" or i =="u":
        vowel+=1
print("Number of vowels are:",vowel)        
```

```python
# #4 Remove Spaces
# Write a program to remove all spaces from a string.
# Input: "a b c d"
# Output: "abcd"

a = "a b c d"
b=""

for i in a:
    if i == " ":
        continue
    else:
        b+=i
        
print(b)        
```

```python
#5 First Non-Repeating Character
# Write a program to find the first non-repeating character in a string.
# Input: "aabbcde"
# Output: "c"

a = "aabbcde"
b = ""
for i in a:
    if i not in b:
        b+=i
        
        

print(b)    
```

```python
# #6 Count Character Frequency
# Write a program to count the occurrence of each character in a string.
# Input: "aabbbcc"
# Output: {'a':2, 'b':3, 'c':2}

a="aabbbcc"
c_a=0
c_b=0
c_c=0

for i in a:
    if i=="a":
        c_a+=1
    elif i=="b":
        c_b+=1
    elif i=="c":
        c_c+=1
output = {"a":c_a,"b":c_b,"c":c_c}       
print(output) 
```

```python
# #7 Check Anagram
# Write a program to check if two strings are anagrams of each other.
# Input: "listen", "silent"
# Output: True

a="listen"
b="silent"
correct=True

if len(a)!=len(b):
    print("Not a anagram")

elif len(a)==len(b):
    for i in a:
           if i not in b:
               correct = False
               
if correct==True:
    print("Anagram")                                     
elif correct==False:
    print("Not a Anagram")                                     
            
```

```python
# #8 Replace Vowels with '*'
# Write a program to replace all vowels in a string with '*'.
# Input: "hello world"
# Output: "h*ll w*rld"

a="hello world"
b=""

for i in range(len(a)):
    if a[i] in "aeiou":
        continue
    else:    
     b = a[i]+"*"
     print(b,end="")
     
```

```python
# #33 Find All Substrings
# Write a program to generate all possible substrings of a string.
# Input: "abc"
# Output: ['a','b','c','ab','bc','abc']
a = "abc"
b = list()

for i in range(len(a)):
    b.append(a[i])
    for j in range(0,len(a)-1):
        if a[j]+a[j+1] not in b :
         b.append(a[j]+a[j+1])
b.append(a)         
print(b)
```

```python
# #34 Sort List of Tuples
# Write a program to sort a list of tuples based on the second element.
# Input: [(1,3),(2,1),(4,2)]
# Output: [(2,1),(4,2),(1,3)]

a = [(1,3),(2,1),(4,2)]

n = len(a)

for i in range(n):
    min_index = i
    
    for j in range(i+1, n):
        if a[j][1] < a[min_index][1]: 
            min_index = j
    a[i], a[min_index] = a[min_index], a[i]

print(a)

    
```
