```python
# 22. Write a Python program to check if a set is a superset of another.
# Input: `{1, 2, 3}` is superset of `{2, 3}` → Output: `True`

a = {1,2,3}
b = {2,3}

for i in b:
    if i in a:
        print(True)
        break
    else:
        print(False)    
        break
 
    
```
