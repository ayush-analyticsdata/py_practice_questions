```python
# #7 Factorial using Function

# Write a function to find factorial of a number.
# Input: 5
# Output: 120

def Factorial(n):
    f = 1
    for i in range(1,n+1):
        f*=i
    print(f"{f} is factorial of {n}")    
Factorial(5)
```

```python
# #11 Prime Number Function

# Write a function to check if a number is prime.
# Input: 7
# Output: Prime

def prime(n):
    for i in range(2,n):
        if n%i==0:
            print(f"{n} is not prime")
            break
        else:
            print(f"{n} is prime")
            break
prime(17)
prime(8)
prime(27)
```

```python
# #12 Fibonacci using Function

# Write a function to print Fibonacci series up to n terms.
# Input: 5
# Output: 0 1 1 2 3

def fibo(n):
    f0 = 0
    f1 = 1
    for i in range(n):
        print(f0,end=" ")
        fn = f0+f1
        f0 = f1
        f1 = fn


fibo(5)       
```

```python
#13 Sum of Digits

# Write a function to find sum of digits of a number.
# Input: 123
# Output: 6

def sum_of_digits(num):
    total = 0
    
    while num > 0:
        digit = num % 10      # Get last digit
        total += digit        # Add digit to total
        num = num // 10       # Remove last digit
        
    return total

# Input
n = 123

# Function call
print("Sum of digits =", sum_of_digits(n))
```

```python
# #14 Palindrome Function

# Write a function to check if a string is palindrome.
# Input: "madam"
# Output: True
def palindrome(s):
    new = ""
    ans=True
    for i in s:
        new = i+new
    if s!=new:
        ans = False
    else:
        pass
    print(ans)    
palindrome("madam") 
```

```python
# #19 Recursive Factorial

# Write a recursive function to find factorial.
# Input: 4
# Output: 24

def facto(n):
    if n==1 or n==0:
        return 1
    else:
        return n*facto(n-1)
facto(6)    
```

```python
# #20 Recursive Fibonacci

# Write a recursive function to generate Fibonacci series.
# Input: 5
# Output: 0 1 1 2 3

def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
n = 5         
for i in range(n):
   print(fibonacci(i), end=" ")   
```

```python
# #25 Function Decorator (Basic)
# Write a decorator function to print "Start" and "End" around function execution

def my_decorator(func):
    def wrapper():
        print("Start")
        func()
        print("End")
    return wrapper
@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```
