```python
# 1. Print numbers from 1 to 50. For each number:

#    * If divisible by 3 → print "Fizz"
#    * If divisible by 5 → print "Buzz"
#    * If divisible by both → print "FizzBuzz"

for i in range(1,51):

  if(i%3==0 and i%5==0):
    print(i,"FizzBuzz")
  elif(i%3==0):
    print(i,"Fizz")
  elif(i%5==0):
    print(i,"Buzz")
```

```python
# 2. Take a number n and print whether each number from 1 to n is even or odd.

n=int(input("Enter a number: "))

for i in range(1,n+1):
  if i%2==0:
    print(i,"Even")
  else:
    print(i,"Odd")
```

```python
# 3. Print all prime numbers between 1 to 100.
# prime number are numbers divisible by 1 and itself

for num in range(1,101):
  if num>1:
    for i in range(2,num):
      if num%i==0:
        break
    else:
     print(num)

```

```python
# 4. Take a number and count how many digits are even and how many are odd.

n=int(input("Enter any number: "))
count_even=0
count_odd=0

while n > 0:
  digit = n%10
  n = n//10
  if digit%2==0:
    count_even+=1
  else:
    count_odd+=1

print("Even digits are: ",count_even)
print("Odd digits are: ",count_odd)


```

```python
# 5. Print multiplication table of a number using for loop and if condition to highlight multiples of 5.
n=int(input("Enter a number for table"))

for i in range(1,11):
  c=i*n
  if(c%5==0):
    print(c,"Divisible by 5")
  else:
    print(c)
```

```python
# 6. Reverse a number using a while loop.

n=int(input("Enter any number:"))
reverse_num =0
while n!=0:
  digit = n%10
  reverse_num = reverse_num*10 + digit
  n = n//10
print("Reversed number:",reverse_num)
```

```python
# 7. Check whether a number is palindrome using loop and condition.

num=int(input("Enter a number: "))
orignal = num
rev = 0

while num!=0:
  digit = num%10
  rev = rev*10 + digit
  num = num//10


if rev==orignal:
  print(orignal,"is a palindrome")
else:
  print(orignal,"is not a palindrome")
```

```python
# 8. Find the sum of all numbers from 1 to n, but skip numbers divisible by 7.
n=int(input("Enter a number till you want to find sum:"))
sum=0
for i in range(1,n+1):
  if(i%7==0):
    continue
  else:
    sum=sum+i
print(f"Sum of all numbers from 1 to {n} but divible by 7 is {sum}")
```
